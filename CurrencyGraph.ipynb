```python
from pycoingecko import CoinGeckoAPI
import pandas as pd
import datetime
import matplotlib.pyplot as plt
import mplfinance as mpf
```

```python
cg = CoinGeckoAPI()
```

```python
#gets current price of bitcoin in the form of ethereum
data = cg.get_price(ids='bitcoin', vs_currencies='ETH')
print("Bitcoin current price in form of ethereum currency")
print(data)
```

```python
class data:
    def gdata(self):
        a=cg.get_coin_ohlc_by_id(id='bitcoin', vs_currency = "usd",days=14)
        return a
    def crea_df(self,a):
        k=[]
        for i in a:
            s=datetime.datetime.fromtimestamp(i[0] / 1e3)
            i[0]=s.date()
        df = pd.DataFrame(a[1:],columns=a[0])
        df.columns=['date','open','high','low','close']
        df.index = pd.DatetimeIndex(df['date'])
        return df

def crea_graph(df):
    mpf.plot(df, type='candle', style='charles',title='BTC',ylabel='price',ylabel_lower='',mav=(2,4,6))
```

```python
d=data()
gd=d.gdata()
dframe=d.crea_df(gd)
crea_graph (dframe)
```
